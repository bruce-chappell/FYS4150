```python
vals = np.load('parta_BinVals.npz')
params = np.load('parta_BinParams.npz')
print(vals.files)
print(params.files)
```

```python
Xsteps = params['binSteps']
plotVals = vals['final_dist']
plotError = vals['error_vals']
totalVals = vals['all_vals']
```

```python
new1 = totalVals[:125]
new2 = totalVals[125:250]
new3 = totalVals[250:375]
new4 = totalVals[375:500]
new5 = totalVals[500:625]
new6 = totalVals[625:750]
new7 = totalVals[750:875]
new8 = totalVals[875:1000]
new = np.array(list(zip(new1,new2,new3,new4,new5,new6,new7,new8)))
new = new.reshape(1000,1000)
# try with 1e5 MC and 1e4 transactinos
```

```python
MCsteps = totalVals.shape[0]
diff_vec = np.zeros(MCsteps-1)
for i in range(0, MCsteps-1):
    diff_vec[i]=np.linalg.norm(new[i]/(i+1)-new[i+1]/(i+2))
```

```python
fig = plt.figure()
plt.plot(Xsteps, plotVals)
plt.show()
```

```python
fig = plt.figure()
plt.loglog(Xsteps[:-1], diff_vec)

plt.show()
```
